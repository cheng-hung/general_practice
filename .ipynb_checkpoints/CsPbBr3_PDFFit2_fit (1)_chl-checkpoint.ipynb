```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-

# %cd /Users/matthewgreenberg/Desktop/PDFFitTest
import pymatgen as pm
from diffpy.pdffit2 import PdfFit
from diffpy.Structure import loadStructure
from pymatgen.io.cif import CifWriter
from pymatgen.io.cif import CifParser


# Specify the path to your .cif file
# cif_file_path = "/Users/matthewgreenberg/Desktop/PDFFitTest/CsPbBr3_Orthorhombic.cif"
cif_file_path = '/Users/chenghunglin/Documents/Git_BNL/general_practice/CsPbBr3_Orthorhombic.cif'

# Initialize the CifParser with the path to your .cif file
parser = CifParser(cif_file_path)

# Parse the .cif file
structure = parser.get_structures()[0] 
structure.remove_oxidation_states()
print(structure)
w = CifWriter(structure,symprec=0.1)
w.write_file('CsPbBr3_pym.cif')

# Create new PDF calculator object.
pf = PdfFit()

# Load data ------------------------------------------------------------------
stru = loadStructure('CsPbBr3_pym.cif')
stru.Uisoequiv = 0.04
# Load experimental x-ray PDF data
qmax = 20.0  # Q-cutoff used in PDF calculation in 1/A
qdamp = 0.03 # instrument Q-resolution factor, responsible for PDF decay
pf.read_data('CsPbBr3.gr', 'X', qmax, qdamp)
#Add loaded .cif 
pf.add_structure(stru)

# Configure Refinement -------------------------------------------------------

# Refine lattice parameters a, b, c.
pf.constrain(pf.lat(1), "@1")
pf.constrain(pf.lat(2), "@2")
pf.constrain(pf.lat(3), "@3")
# set initial value of parameter @1, @2, @3
pf.setpar(1, pf.lat(1))
pf.setpar(2, pf.lat(2))
pf.setpar(3, pf.lat(3))

# Refine phase scale factor.  Right side can have formulas.
pf.constrain('pscale', '@20 * 2')
pf.setpar(20, pf.getvar(pf.pscale) / 2.0)

# Refine PDF damping due to instrument Q-resolution.
# Left side can be also passed as a reference to PdfFit object
pf.constrain(pf.qdamp, '@21')
pf.setpar(21, 0.03)

# Refine sharpening factor for correlated motion of close atoms.
pf.constrain(pf.delta2, 22)
pf.setpar(22, 0.0003)

# Set temperature factors isotropic to each atom with an initial value 0.04
# idx starts from 1 not 0
# 1-4: Cs
for idx in range(1, 5):
    pf.constrain(pf.u11(idx), '@11')
    pf.constrain(pf.u22(idx), '@11')
    pf.constrain(pf.u33(idx), '@11')
pf.setpar(11, 0.04)

# 5-8: Pb
for idx in range(5, 9):
    pf.constrain(pf.u11(idx), '@22')
    pf.constrain(pf.u22(idx), '@22')
    pf.constrain(pf.u33(idx), '@22')
pf.setpar(22, 0.04)

# 9-16: Br
for idx in range(9, 17):
    pf.constrain(pf.u11(idx), '@33')
    pf.constrain(pf.u22(idx), '@33')
    pf.constrain(pf.u33(idx), '@33')
pf.setpar(33, 0.04)

# 16-20: Br
for idx in range(17, 21):
    pf.constrain(pf.u11(idx), '@34')
    pf.constrain(pf.u22(idx), '@34')
    pf.constrain(pf.u33(idx), '@34')
pf.setpar(34, 0.04)

# Refine 

pf.pdfrange(1, 1.5, 19.99)
pf.refine()

# Save results
pf.save_pdf(1, "CsPbBr3_refinement.fgr")
pf.save_struct(1, "CsPbBr3_refinement.rstr")
pf.save_res("CsPbBr3_refinement.res")

# Plot results

# pylab is matplotlib interface with MATLAB-like plotting commands
import pylab

# obtain data from PdfFit calculator object
r = pf.getR()
Gobs = pf.getpdf_obs()
Gfit = pf.getpdf_fit()

# calculate difference curve, with pylab arrays it can be done
# without for loop
Gdiff = pylab.array(Gobs) - pylab.array(Gfit)
Gdiff_baseline = -1

pylab.plot(r, Gobs, 'ko')
pylab.plot(r, Gfit, 'b-')
pylab.plot(r, Gdiff + Gdiff_baseline, 'r-')

pylab.xlabel(u'r ()')
pylab.ylabel(u'G ($^{-2}$)')
pylab.title('Fit of CsPbBr3 to x-ray experimental PDF')

# display plot window, this must be the last command in the script
pylab.show()
```
